Your next question is probably, how do I deal with these columns with missing values?

There are a few ways to deal with it:
1. Find out the actual value and fill it in. This is always the best solution.
2. Drop the column entirely if it's not essential to your analysis.
3. Keep the column, especially if it's critical to your analysis. In this case, the director, cast, and country columns are important, so it's best to keep them, and consider imputation, which means replacing missing data with substituted values. However, it's not possible to do this in most cases when the data are string values and not numerical values.

Your next next question is probably, how do I deal with these rows with missing values?
In some cases, you might want to examine the distribution of missing values across all the rows of your dataset (given that your dataset doesn’t have a large number of observations/rows). From here, you can choose from the choices above depending on how important the rows are to your analysis. For instance, your dataset contains recorded data of something that is changing over time. Even though a row can contain missing values, you might not want to eliminate it because there is important time information you want to retain.

note: NaN means Not a Number in pandas. we need to replace them

For string type values, we can replace NaN values
with “” or “None” or any string that can indicate to you that
there isn’t any value in that entry

```python
# getting all the columns with string as value

str_cols = list(df.columns)
str_cols.remove('release_year') #release year is int not string and we won't use it today.

# removing leading and trailing characters from columns with str type
for i in str_cols:
    df[i] = df[i].str.strip()
    

# names of the columns to replace NaN values:
columns = ['director', 'cast', 'country', 'rating', 'date_added']

# looping through the columns to fill the entries with NaN values with ""
for column in columns:
    df[column] = df[column].fillna("")
```

```python
df
```

```python
####
from datetime import datetime

# examining rows with null values for date_added column
rows = []
for i in range(len(df)): # same as for i in range number_of_rows (we get the num of rows with len() on the df)
    if df['date_added'].iloc[i] == "": # iloc function : go to the i'th place in the column and fetch the data.
        rows.append(i)
        
# extracting months added and years added
month_added = []
year_added = []

for i in range(len(df)):
    # replacing NaN values with 0
    if i in rows:
        month_added.append(0)
        year_added.append(0)
    else:
        date = df['date_added'].iloc[i].split(" ")
        month_added.append(date[0])
        year_added.append(int(date[2]))
        
# turning month names into month numbers
for i, month in enumerate(month_added):
    if month != 0:
        datetime_obj = datetime.strptime(month, "%B")
        month_number = datetime_obj.month
        month_added[i] = month_number
        
# checking all months
print(set(month_added))
print(set(year_added))

# inserting the month and year columns into the dataset
df.insert(7, "month_added", month_added, allow_duplicates = True)
df.insert(8, "year_added", year_added, allow_duplicates = True)
df.head()
```

We will deal with it by separating the type of content into 2 different datasets and naturally, the duration column will just be numerical and just have 1 type of time unit. This way, you can easily and clearly plot using the values.

```python
# separating original dataset to tv show and movie dataset#

# creating dataframe for tv shows
netflix_shows = df[df['type'] == "TV Show"]
# changing the indexing so it will start from 0
netflix_shows = netflix_shows.set_index([pd.Index(range(0, len(netflix_shows)))])


# creating dataframe for movies
netflix_films = df[df['type'] == "Movie"]
# changing the indexing so it will start from 0
netflix_films = netflix_films.set_index([pd.Index(range(0, len(netflix_films)))])
```

We also have to create a function to extract the number from that column so that it can be inserted into the columns of the 2 new dataset, because the duration column has both strings and numbers - and now we don't need the string part that tells that what the units are (min and seasons)

```python
# get length of movie or number of seasons of show
def getDuration(data):
    count = 0
    durations = []
    for value in data:
    # filling in missing values
        if type(value) is float:
            durations.append(0)
        else:
            values = value.split(" ")
            durations.append(int(values[0]))
    return durations
    
# inserting new duration type column for shows (renamed column)
netflix_shows.insert(11, 'seasons', getDuration(netflix_shows['duration']))
netflix_shows = netflix_shows.drop(['duration'], axis = 1)
netflix_shows.head()

# inserting new duration type column for films (renamed column)
netflix_films.insert(11, 'length', getDuration(netflix_films['duration']))
netflix_films = netflix_films.drop(['duration'], axis = 1)
netflix_films.head()
```

let’s look at the country and listed_in column. By now, you must have realized that it’s not as easy as the rating column to extract unique values. This is because the values in those columns are words joined together by commas, making it more difficult to extract the set of words and then find unique words from that set.
we will need to  implement a unique function for this special case.

To start, let’s think about what data structure can give us unique values easily.


If you guessed sets, you’re right! Given its ability to store unique elements of the same type in sorted order, it’s a fitting data structure for what we want to do.

```python

# function to get unique values of a column
def getUnique(data):
    unique_values = set()
    for value in data:
        if type(value) is float:
            unique_values.add(None)
        else:
            values = value.split(", ")
            for i in values:
                unique_values.add(i)
    return list(unique_values)
```

```python

# getting unique country names
unique_countries = getUnique(df['country'])
unique_countries
```

Next, let’s examine the list of unique countries to see if there are any inconsistencies or mistakes. By doing so and with a little bit of Googling, we can see there are some issues with this list:

1. There’s both the Soviet Union and Russia
2. There’s both the West/East Germany and Germany

We can easily fix this with a few modifications to the dataset.

```python

# converting soviet union to russia and east/west germany to germany
for i in range(len(df)):
    if type(df['country'].iloc[i]) is not float:
        countries = df['country'].iloc[i].split(", ")
        for j in range(len(countries)):
            if "Germany" in countries[j]:
                countries[j] = "Germany"
            elif "Soviet Union" in countries[j]:
                countries[j] = "Russia"
        df['country'].iloc[i] = ", ".join(countries)
```

In both the TV shows and films dataset, there is a “TV Shows” and “Movies” genres. Technically, this isn’t a genre but could be a label of the type of content. To confirm this, we should print out the counts of these “genres” appearing in the respective datasets.

The hypothesis is that if these “genres” appear in all the rows of the datasets, it means that they’re simply labels. Otherwise, we’ll have to investigate further as to what those “genres” represent.

```python

# checking for TV shows
# replace netflix_shows with netflix_films to check for movies
count = 0
index = []
for i, value in enumerate(netflix_shows['listed_in']):
    genres = value.split(", ")
    if "TV Shows" in genres:
        count += 1
        index.append(i)
print("count %s" %count)
print("index %s" %index)
```

As the count of the “genres” is less than the size of the datasets, let’s use the output of the code to examine the rows.

Since I’ve written the code to specifically output the row indices in a list, we can easily use that list and the iloc function to get a view of the rows.

```python
# printing the first 5 rows of all rows that have TV Shows as its genre
netflix_shows.iloc[index[0:5]]

# printing the first 5 rows of all rows that have Movies as its genre
netflix_films.iloc[index[0:5]]
```

Taking a look at the rows, it is now obvious that the “TV Shows” and “Movies” genre was used to signify that these content didn’t have a genre in the first place. Now that we understand what this meant, we can either choose to exclude or include it in our analysis
